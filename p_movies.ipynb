```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Assuming you have loaded your DataFrame into a variable called 'df'

# Drop rows with missing values in 'Director', 'Actor 1', 'Actor 2', 'Actor 3', and 'Genre' columns
df.dropna(subset=['Director', 'Actor 1', 'Actor 2', 'Actor 3', 'Genre'], inplace=True)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Encode 'Director'
df['Director_encoded'] = label_encoder.fit_transform(df['Director'])

# Encode 'Actor 1', 'Actor 2', 'Actor 3' separately
df['Actor1_encoded'] = label_encoder.fit_transform(df['Actor 1'])
df['Actor2_encoded'] = label_encoder.fit_transform(df['Actor 2'])
df['Actor3_encoded'] = label_encoder.fit_transform(df['Actor 3'])

# For encoding 'Genre', you can use one-hot encoding
# First, split the genres into separate columns
genre_dummies = df['Genre'].str.get_dummies(sep=', ')

# Concatenate the one-hot encoded genres with the original DataFrame
df = pd.concat([df, genre_dummies], axis=1)

# Drop the original categorical columns
df.drop(['Genre', 'Director', 'Actor 1', 'Actor 2', 'Actor 3'], axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df.dropna(subset="Rating",inplace=True)
```

```python
df["Votes"].fillna(-1,inplace=True)
df["Duration"].fillna("-1 min",inplace=True)
```

```python

df['Duration_minutes'] = df["Duration"].str.extract('(\d+)').astype(int)
```

```python
df['Year_int'] = df['Year'].str.strip('()').astype(int)
```

```python
df['Vote'] = df['Votes'].str.replace(',', '').astype(int)
df=df.drop(columns=["Name","Duration","Year","Votes","Reality-TV"])
```

```python
import pandas as pd

# Calculate correlation coefficients with the 'Rating' column
correlation = df.corr()['Rating'].abs()

# Drop columns with correlation less than 0.3
columns_to_drop = correlation[correlation < 0.02].index
df.drop(columns=columns_to_drop, inplace=True)

# Now, your DataFrame contains only columns with correlation coefficient >= 0.3 with the 'Rating' column
```

```python
x=df.drop(columns=['Rating'])
y=df['Rating']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Splitting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initializing and training the linear regression model
model=RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
accuracy=r2_score(y_test,y_pred)
accuracy
```
